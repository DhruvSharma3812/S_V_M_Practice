```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
print(dir(iris))
print(iris.feature_names)
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df.sample(5,random_state=42)
```

```python
df['target'] = iris.target
df.head()
```

```python
iris.target_names
```

```python
df[df.target==1].head()
```

```python
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])
df.sample(14)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df0 = df[df.target==0]
df1 = df[df.target==1]
df2 = df[df.target==2]
```

```python
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],color='red',marker='d',alpha=0.5,label='setosa')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color='blue',marker='o',alpha=0.5, label='versicolor')
plt.scatter(df2['sepal length (cm)'],df2['sepal width (cm)'],color='green',marker='s',alpha=0.5, label='virginica')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Dimensions by Class')
plt.legend()
plt.show()
```

```python
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],color='red',marker='d',alpha=0.5,label='setosa')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color='blue',marker='o',alpha=0.5, label='versicolor')
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'],color='green',marker='s',alpha=0.5, label='virginica')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Petal Dimensions by Class')
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['target','flower_name'],axis='columns')
X.head()
```

```python
y = df.target
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
len(X_test)
```

```python
from sklearn.svm import SVC
model=SVC(C=5)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
