```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
print(dir(digits))
```

```python
print(digits.feature_names)
digits.data.shape
```

```python
df = pd.DataFrame(digits.data,columns=digits.feature_names)
df
```

```python
df['target'] = digits.target
df['target'].sample(20)
```

```python
df[df.target==9]
```

```python
X = df.drop('target',axis=1)
y = df['target']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
a=pd.Series(y_pred)
a.size
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
figure,axes=plt.subplots(1,2,figsize=(16,7))

y_test_bin_count = np.bincount(y_test)
y_pred_bin_count = np.bincount(y_pred)

axes[0].bar(range(len(y_test_bin_count)), y_test_bin_count, color='skyblue')
for i,count in enumerate(y_test_bin_count):
    axes[0].text(i,count-10,str(count),ha='center',va='bottom',color='white',fontweight='bold',fontsize=20)
axes[0].set_xticks(np.arange(10))
axes[0].set_title("Real")

axes[1].bar(range(len(y_pred_bin_count)),y_pred_bin_count,color='salmon') 
for i,count in enumerate(y_pred_bin_count):
    axes[1].text(i,count-10,str(count),ha='center',va='bottom',color='white',fontweight='bold',fontsize=20)
axes[1].set_xticks(np.arange(10))
axes[1].set_title("Predicted")

plt.suptitle('Digits Recognization through SVM')
plt.tight_layout()
plt.show()
```
