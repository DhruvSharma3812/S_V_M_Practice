```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
print(dir(digits))
```

```python
print(digits.feature_names)
digits.data.shape
```

```python
df = pd.DataFrame(digits.data,columns=digits.feature_names)
df
```

```python
df['target'] = digits.target
df['target'].sample(20)
```

```python
df[df.target==9]
```

```python
X = df.drop('target',axis=1)
y = df['target']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
a=pd.Series(y_pred)
a.size
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
figure,axes=plt.subplots(1,2,figsize=(16,7))

y_test_bin_count = np.bincount(y_test)
y_pred_bin_count = np.bincount(y_pred)

axes[0].bar(range(len(y_test_bin_count)), y_test_bin_count, color='skyblue')
for i,count in enumerate(y_test_bin_count):
    axes[0].text(i,count-10,str(count),ha='center',va='bottom',color='white',fontweight='bold',fontsize=20)
axes[0].set_xticks(np.arange(10))
axes[0].set_title("Real")

axes[1].bar(range(len(y_pred_bin_count)),y_pred_bin_count,color='salmon')
for i,count in enumerate(y_pred_bin_count):
    axes[1].text(i,count-10,str(count),ha='center',va='bottom',color='white',fontweight='bold',fontsize=20)
axes[1].set_xticks(np.arange(10))
axes[1].set_title("Predicted")

plt.suptitle('Digits Recognization through SVM')
plt.tight_layout()
plt.show()
```

```python
import csv
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

num_images = len(data) - 1

pixel_values = []

for i in range(1, len(data)): 
    image = data[i]["image"]

    image = image.convert("L")
    image_resized = image.resize((8, 8), Image.Resampling.LANCZOS)

    image_array = np.array(image_resized).astype(float) / 255.0

    inverted_image_array = 1.0 - image_array

    image_array_scaled = inverted_image_array * 16.0

    image_array_rounded = np.round(image_array_scaled, 2) 

    print(f"Image {i} - 8x8 floating point pixel values (scaled to 0-16):")
    print(image_array_rounded)
    print()

    pixel_values.append(image_array_rounded.flatten())

    plt.subplot(2, 5, i)
    plt.imshow(image_resized, cmap='hot')
    plt.title(f"Image {i}", fontsize=8)
    plt.axis("off")

csv_file_path = 'pixel_values_scaled.csv'

with open(csv_file_path, mode='w', newline='') as file:
    writer = csv.writer(file)
    header = [f"pixel_{i}_{j}" for i in range(8) for j in range(8)]
    writer.writerow(header)
    writer.writerows(pixel_values)

plt.tight_layout()
plt.show()
```
